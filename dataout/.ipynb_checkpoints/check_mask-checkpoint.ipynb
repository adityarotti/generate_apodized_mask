```python
import healpy as h
```

```python
md=h.read_map("diffuse_emission_mask_fh1a_apo120a.fits")
mp=h.read_map("peaked_emission_mask_fh1a_apo120a.fits")

mdo=h.read_map("../datain/diffuse_emission_mask.fits")
mpo=h.read_map("../datain/peaked_emission_mask.fits")
```

```python
h.orthview(mp,rot=(0,90))
h.orthview(mpo,rot=(0,90))
h.orthview(md,rot=(0,90))
h.orthview(mdo,rot=(0,90))
```

```python
clmd=h.alm2cl(h.map2alm(md,lmax=1024))
clmp=h.alm2cl(h.map2alm(mp,lmax=1024))
clmdo=h.alm2cl(h.map2alm(mdo,lmax=1024))
clmpo=h.alm2cl(h.map2alm(mpo,lmax=1024))
```

```python
plot(clmd,label="Diffuse")
plot(clmdo,label="Diffuse o")
plot(clmp,label="Peaked")
plot(clmpo,label="Peaked o")
legend()
loglog()
```
